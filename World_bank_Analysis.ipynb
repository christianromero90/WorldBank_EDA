## INTRODUCTION

### North America and Latin America: Same Continent, Yet so Different
<font size = 3>
The American continent landmass has two geographical regions: North America and South America. The way in which these regions are separated is by the Isthmus of Panama that is located in what we also call Central America. However, due to strong differences in cultures and languages, there is another way to group these countries. All countries from Mexico to the south are denominated as Latin American, while The United States and Canada are grouped as Anglo-America.

![continent_america](./images/continent_america.png)

The Anglo-American countries are known for being stablished as first world ones with powerful governments. The United States itself is globally recognized by its large economy that has called the attention of foreigners to move to the country to pursue “The American Dream” (1). Another popular characteristic is its protagonist on the musical and film industry, by being home of many famous artist and actresses such as Elvis Presley, Marilyn Monroe, and Michael Jackson. Lastly, they are also home for global recognized food and beverage brands, such as Coca Cola and McDonalds. Given this northern region colonization history, USA and Canada have English as their official language, with the difference that Canada also has some areas where French dominates as the main language.

Latin American countries in the other hand, are mainly known for the infamous sport, soccer (2). Brazil and Argentina are the leading ones who have represented the region by being champions several times in the World Cup. Given that they were colonized by the Spanish, the culture and the people from this region is a mix of Indigenous, European, and African descendance (2). The majority have adopted Spanish as their main language and based their lifestyle on the Roman Catholic religion.  For this reason, Latin American families tend to be large ones and creating strong ties with each other’s. Lastly, these countries are known for their tropical essence, beaches, music, and tasty food is something that calls the attention from tourists. 

This background is important to understand the analysis that we will present in this report. We used a dataset that shows around 40 different dimensions for all north American and central Latin American countries. Our objective with the report is to find a country that would best represent both regions. The attributes used to define the country’s representation are variables under the economic, health, environmental, education, and social sectors. After organizing and testing our data, we have decided that Mexico is the one best representing both regions. More about our decision will be explained in our analysis.

```python
# Defining colors for MEAN & MEDIAN
mean_color='#B815CE'
median_color = '#0D2763'

legend_mean = mpatches.Patch(color=mean_color, label='MEAN')
legend_median = mpatches.Patch(color=median_color, label='MEDIAN')
```

```python

# ENERGY CONSUMPTION BY INCOME LEVEL in Regions of NORTH AMERICA & LATIN AMERICA

"""
First, creating a box plot for the three columns in the enviroment Dataframe and analyze the distribution
For this distribution we are looking at North America-Latin American countries to have more data and better
visualization on the distribution
"""
energy_consumption = filter_data_region.loc[:,"Energy use (kg of oil equivalent) per $1,000 GDP (constant 2011 PPP)"]\
                     [filter_data_region["Income Group"] == "Upper middle income"].dropna()

sns.set(rc={"figure.figsize":(8,6)})
sns.set_style("white")

sns.distplot(a = energy_consumption,
             bins  = 6,
             hist  = True,
             kde   = True,
             rug   = False,
             color = '#2BA4B4')

plt.axvline(x = energy_consumption.mean(),
            color = mean_color)

plt.axvline(x = energy_consumption.median(),
            color = median_color)

plt.title(label = "Distribution of Energy use (kg of oil equivalent) per $1,000 GDP (constant 2011 PPP)")
plt.legend(handles=[legend_mean, legend_median])

plt.show()


"""
For this distribution we are looking at North America-Latin American Upper-Middle Income countries to have more data and better
visualization. The Hot Rod region has only 11 data points and it's difficult to visualize it.
"""
sns.set(rc={"figure.figsize":(8,6)})
sns.set_style("white")

gdp_energy_unit = filter_data_region.loc[:,"GDP per unit of energy use (constant 2011 PPP $ per kg of oil equivalent)"]\
                                           [filter_data_region["Income Group"] == "Upper middle income"].dropna()
sns.distplot(a     = gdp_energy_unit,
             bins  = 7,
             hist  = True,
             kde   = True,
             rug   = False,
             color = '#2BA4B4')

plt.axvline(x = gdp_energy_unit.mean(),
            color = mean_color)


plt.axvline(x = gdp_energy_unit.median(),
            color = median_color)

plt.title("GDP per unit of energy use (constant 2011 PPP $ per kg of oil equivalent)")
plt.legend(handles=[legend_mean, legend_median])
plt.show()
```

```python
#Checking the distribution, the median seem the best value to fill the missing data points as the
#Histogram seems to be positively skewed.

# Replacing Missing Values with MEDIAN as they are closer to normal distribution

median_energy_consumption = energy_consumption.median()
gdp_median = gdp_energy_unit.median()


enviroment_df_hult["GDP per unit of energy use (constant 2011 PPP $ per kg of oil equivalent)"].fillna(value = gdp_median, inplace=True)
enviroment_df_hult["Energy use (kg of oil equivalent) per $1,000 GDP (constant 2011 PPP)"].fillna(value=median_energy_consumption, inplace=True)
```

```python
# -------------------THE ANALYSIS & BETWEEN WORLD & AVERAGE REPRESENTATION OF MEXICO ---------------------------

differences = []
names = []

dfs_world = [enviroment_df_world, education_df_world, health_df_world, social_df_world, economics_df_world]
dfs_hult = [enviroment_df_hult, education_df_hult, health_df_hult, social_df_hult, economics_df_hult]

k = 0
start_index = 4

for x in dfs_world:
    
    data_used_1 = x
    data_used_2 = dfs_hult[k]
    k+=1

    for index in range(start_index, len(data_used_1.columns.tolist())):
        z = data_used_1.columns.tolist()
        i = z[index]

        a = data_used_2[i].mean()
        b = data_used_1[i].mean()

        diff = a-b

        differences.append(round(diff,2))
        names.append(i)
        
df = pd.DataFrame(list(zip(names, differences)), columns =['Name', 'val']) 
df = df.drop(15, axis=0)

fig, ax = plt.subplots(figsize = [10, 15])
grph = sns.barplot( x = "val",
             y = "Name",
             data = df)
plt.xlim(-150,150)
grph.set_title("WHAT MAKES A REGION UNIQUE",fontsize=15)
plt.show()
```
